```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.font_manager as fm
plt.rcParams['font.family'] = 'HYGothic-Extra'  # 폰트 확인
font_names = []
for font in fm.fontManager.ttflist:
    font_names.append(font.name)
font_names.sort()
for font in font_names:
    print(font)
```

```python
# 데이터 불러오기
df = pd.read_csv("국민건강보험공단_제1형 당뇨병 환자 수_20241231.csv", encoding='cp949')
x = [] # 나이
y = [] # 당뇨병 환자 수
a = df.values.tolist()
for i in range(len(a)): # 데이터 잔처리 및 추출
    if a[i][3] == '*':
        a[i][3] = 0
    x.append(int(a[i][1]))
    y.append(int(a[i][3]))

# 환자의 나이(x)와 당뇨병 환자 수(y)를 numpy 배열로 변환
x = np.array(x).reshape(-1, 1)
y = np.array(y)

# 선형 회귀 모델 생성 및 학습
model = LinearRegression()
model.fit(x, y)

# 20살과 40살의 당뇨병 환자 수 예측
ages = np.array([[20], [40]])
predictions = model.predict(ages)
print("20살 당뇨병 환자 수 예측:",predictions[0])
print("40살 당뇨병 환자 수 예측:",predictions[1])
```

```python
# 시각화
y_pred = model.predict(x) # 회귀선 예측값
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='당뇨병 환자')
plt.plot(x, y_pred, color='red', label='회귀선')
plt.title('나이별 당뇨병 환자')
plt.xlabel('나이')
plt.ylabel('환자 수')
plt.legend()
plt.show()
```
