```python
import pandas as pd

df = pd.read_csv("학생건강검사.csv", encoding='cp949')
a = df.values.tolist()

mancount=[0,0,0,0,0,0] # 1~6학년 남자 초등학생 수 mancount[0] = 1학년 남자 초등학생 수
mansum=[0,0,0,0,0,0] # 1~6학년 남자 초등학생 키 합 mansum[0] = 1학년 남자 초등학생 키 합
for i in range(len(a)): 
    if a[i][1] == '초' and a[i][3] == '남': # 전체 학생 중 초등학생, 남자
        mansum[int(a[i][2]) - 1] +=a [i][4] # a[i][2]학년의 키를 masum[학년 - 1]에 더함
        mancount[int(a[i][2]) - 1] += 1 # a[i][2]학년의 남자 초등학생 수를 mancount[학년 - 1]에 더함

for i in range(6):
   print(i + 1,"학년 남자 초등학생의 수는", mancount[i]) # 1~6학년 남자 초등학생 수 출력
for i in range(6):
    print(i + 1,"학년 남자 초등학생의 평균 키는",mansum[i]/mancount[i]) # 1~6학년 남자 초등학생 평균 키 출력
```

```python
womancount=[0,0,0,0,0,0] # 1~6학년 여자 초등학생 수 womancount[0] = 1학년 여자 초등학생 수
womansum=[0,0,0,0,0,0] # 1~6학년 여자 초등학생 키 합 womansum[0] = 1학년 여자 초등학생 키 합
for i in range(len(a)): 
    if a[i][1] == '초' and a[i][3] == '여': # 전체 학생 중 초등학생, 여자
        womansum[int(a[i][2]) - 1] += a[i][4] # a[i][2]학년의 키를 masum[학년 - 1]에 더함
        womancount[int(a[i][2]) - 1] += 1 # a[i][2]학년의 여자 초등학생 수를 mancount[학년 - 1]에 더함

for i in range(6):
   print(i + 1,"학년 여자 초등학생의 수는", womancount[i]) # 1~6학년 여자 초등학생 수 출력
for i in range(6):
    print(i + 1,"학년 여자 초등학생의 평균 키는",womansum[i]/womancount[i]) # 1~6학년 여자 초등학생 평균 키 출력
```

```python
for i in range(5):
    print(i + 1,"학년", i + 2,"학년의 남자 초등학생 평균 키 차이는",-(mansum[i]/mancount[i]-mansum[i+1]/mancount[i+1])) # 1~6 학년 남자 초등학생 평균 키 차이 출력
for i in range(5):
    print(i + 1,"학년", i + 2,"학년의 여자 초등학생 평균 키 차이는",-(womansum[i]/womancount[i]-womansum[i+1]/womancount[i+1])) # 1~5학년 여자 초등학생 평균 키 차이 출력
print("다음의 값들을 분석해보면 5학년까지의 여자 초등학생의 평균 키가 남자 초등학생의 평균 키보다 크고, 성장 속도도 더 빠르다.")
print("하지만 6학년이 되면 남자 초등학생의 평균 키가 여자 초등학생의 평균 키보다 더 커지고 성장 속도도 더 빠르다")
print("여자 초등학생의 2차 성징이 남자 초등학생의 2차 성징보다 빠르기 때문에 이러한 현상이 나타나는 것으로 보인다.")
```
