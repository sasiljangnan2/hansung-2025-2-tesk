```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.font_manager as fm

df = pd.read_csv("학생건강검사.csv", encoding='cp949')
a = df.values.tolist()

mancount=[0,0,0] # 초중고 학생 수 count[0] = 초등학생 남자 수
wamoncount=[0,0,0] # 초중고 학생 수 count[0] = 초등학생 여자 수
mansum=[0,0,0,0,0,0] # 1~6학년 남자 초등학생 키 합 mansum[0] = 1학년 남자 초등학생 키 합
wamonsum=[0,0,0,0,0,0] # 1~6학년 여자 초등학생 키 합 wamosum[0] = 1학년 여자 초등학생 키 합
for i in range(len(a)): 
    if a[i][3] == '남': # 전체 학생 중 남자
        if a[i][1] == '초': # 초등학생
            mancount[0] += 1
            mansum[0] += a[i][4] # 남자 초등학생 키 합
        elif a[i][1] == '중': # 중학생
            mancount[1] += 1
            mansum[1] += a[i][4] # 남자 중학생 키 합
        elif a[i][1] == '고': # 고등학생
            mancount[2] += 1
            mansum[2] += a[i][4] # 남자 고등학생 키 합
    elif a[i][3] == '여': # 전체 학생 중 여자
        if a[i][1] == '초': # 초등학생
            wamoncount[0] += 1
            wamonsum[0] += a[i][4] # 여자 초등학생 키 합 
        elif a[i][1] == '중': # 중학생
            wamoncount[1] += 1
            wamonsum[1] += a[i][4] # 여자 중학생 키 합
        elif a[i][1] == '고': # 고등학생
            wamoncount[2] += 1
            wamonsum[2] += a[i][4] # 여자 고등학생 키 합
sum=[0,0,0,0,0,0] # 1~6학년 남자 초등학생 키 합 mansum[0] = 1학년 남자 초등학생 키 합

print("남자 초등학생의 평균 키: ", mansum[0]/mancount[0])
print("남자 중학생의 평균 키: ", mansum[1]/mancount[1])
print("남자 고등학생의 평균 키: ", mansum[2]/mancount[2])
print("여자 초등학생의 평균 키: ", wamonsum[0]/wamoncount[0])
print("여자 중학생의 평균 키: ", wamonsum[1]/wamoncount[1])
print("여자 고등학생의 평균 키: ", wamonsum[2]/wamoncount[2])
```

```python
plt.rcParams['font.family'] = 'HYGothic-Extra'  # 폰트 확인
font_names = []
for font in fm.fontManager.ttflist:
    font_names.append(font.name)
font_names.sort()
for font in font_names:
    print(font)
```

```python
X = [ "초등학생", "중학생", "고등학생"] # x축 값
Y = [mansum[0]/mancount[0], mansum[1]/mancount[1], mansum[2]/mancount[2]] # y축 값 남자 학생의 평균 키
Y2 = [wamonsum[0]/wamoncount[0], wamonsum[1]/wamoncount[1], wamonsum[2]/wamoncount[2]] # y축 값 여자 학생의 평균 키
plt.bar(X, Y, width= 0.7, color = "blue" ,label="남자")	# 남자 선 그래프
plt.bar(X, Y2, width=  0.7, color = "red", label="여자",  bottom = Y)	# 여자 선 그래프
plt.title("초등학생, 중학생, 고등학생의 평균 키") # 그래프 제목
plt.xlabel("학교급 별 키")	# x축 제목
plt.ylabel("키") # y축 제목
plt.legend(loc="upper left") # 범례 표시
plt.show()

```
