```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. 입력과 정답 데이터
# x1: 공부 시간, x2: 수면 시간
X = np.array([
    [1, 5],  # 공부 1시간, 수면 5시간
    [2, 4],
    [3, 3],
    [4, 2],
    [5, 1],
])

# y: 시험 점수
y = np.array([60, 65, 70, 75, 80])

# 2. 모델 학습
model = LinearRegression()
model.fit(X, y)

# 3. 회귀식 출력
print("회귀 계수 (기울기):", model.coef_)
print("절편 (intercept):", model.intercept_)

# 4. 예측 예시
new_data = np.array([[3, 4]])  # 공부 3시간, 수면 4시간
predicted_score = model.predict(new_data)
print("예측된 시험 점수:", predicted_score[0])
```
