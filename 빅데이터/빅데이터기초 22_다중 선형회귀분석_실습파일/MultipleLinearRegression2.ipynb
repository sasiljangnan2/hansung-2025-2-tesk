```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. 
X = np.array([
    [1, 5],
    [2, 4],
    [3, 3],
    [4, 2],
    [5, 1],
])
y = np.array([60, 65, 70, 75, 80])

# 2.  
model = LinearRegression()
model.fit(X, y)

# 3.  
new_data = np.array([[3, 4]])
predicted_score = model.predict(new_data)

# 4.    
x1_range = np.linspace(1, 5, 10)
x2_range = np.linspace(1, 5, 10)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)
y_grid = model.coef_[0] * x1_grid + model.coef_[1] * x2_grid + model.intercept_

# 5. 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#   
ax.scatter(X[:, 0], X[:, 1], y, color='red', label='Training Data')

#   
ax.scatter(new_data[:, 0], new_data[:, 1], predicted_score, color='green', s=100, label='Predicted Point')

#  
ax.plot_surface(x1_grid, x2_grid, y_grid, alpha=0.5)

#   ()
ax.set_xlabel('  (x1)')
ax.set_ylabel('  (x2)')
ax.set_zlabel('  (y)')
ax.set_title('   ')

#   
ax.legend()
plt.tight_layout()
plt.show()
```
